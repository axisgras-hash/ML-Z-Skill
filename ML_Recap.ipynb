# Step 4: Creating ML Model

```python
X = df[['year']]
y = df['per capita income (US$)']
```

```python
X.shape
```

```python
y.shape
```

```python
df.head()
```

```python
# train_test_split: random_state:

# Hello Everyone, we are learning ML
# are we Hello Everyone
```

```python
# model: just a mathematical algo that learns from the given dataset, to proper
# understand the pattern
# First:  custom mathematical Function:
# SKlearn:
```

```python
model = LinearRegression()
```

```python
# training part
model.fit(X,y)
# y =  mx + c
# m: slope
# x: feature
# c: intercept: when x is zero y = c
# give random value of m, c
# gradient descent: Minize the Loss by adjusting m, c
```

```python
y_pred = model.predict(X)
print(y_pred)
```

```python
plt.title('Yearly PCI')
plt.scatter(df['year'], df['per capita income (US$)'], color = 'r', marker = '.', label = 'actual data')
plt.plot(df['year'], y_pred , color = 'b', label = 'best fit line')
plt.legend()
plt.show()
```

```python
# residuals, loss, cost
# loss = predicted- actual
# mse, mae, rmse

df['y_pred'] = y_pred
```

```python
df
```

```python
# mae: mean absolute error
#
(df['per capita income (US$)'] - df['y_pred']).mean()
```

```python
mean_absolute_error(y,y_pred)
```

```python
mse = mean_squared_error(y, y_pred)
mse
```

```python
rmse = root_mean_squared_error(y, y_pred)
rmse
```
