# Overfitting, underfitting and Boosting

```python
# Overfitting: It is a problem in ML where our model trained well on training
# dataset or giving good accuracy or score on training dataset but not
# performing well with testing dataset
```

```python
# Undefitting: Model not performing well either on training dataset or on testing
# dataset
# Model not trained properly
```

```python
# How to Eliminate Undefit and Overfitt  Problems
```

```python
# Solutions
# 1. Regularization(L1, L2)
# 2. Cross- Validation (cv)
# 3. Early Stopping
# 4. Drop out: Drop neuron
# 5. you can try multi-model concept
```

```python
# Balanced fit:
# It performing well on both training as well as testing dataset
```

# Regularization(L1, L2)

```python
# Linear regression:
  # L1 Model Regularization: Lasso Regression
  # It prevent from Overfitting
  # y = W.x + b
  # mae, mse, rmse error:
  # mae: mean absolute error
  # mae = mean(|y_actual - y_pred|)
  # L1 technique: lyapp more penalty to large missed calculation
  # L1: penalize large Weights:
  # LKG students : broke car window
  # 12th Passout : broke car window
  # loss formula in L1
  # Loss = loss_original + lambda_sum|w|  (Extra)
```

```python
# L2: Regularization(Ridge Regression)
```

```python
 # Loss = loss_original + lambda_sum|w**2|  (Extra)
 # Whenever you are getting multicollinearity problem
 # one feature is dependent on other features
 # It eliminate the overfitting issue
```

```python
# Elastic Net( combination of both L1 + L2 technique)
```

```python
# L1, L2, Elastic Net(L1 + L2)
# Elastic Net(L1 + L2)
# New_Loss = loss_original + lambda_1_sum|w|  + lambda_2_sum|w**2|
```
